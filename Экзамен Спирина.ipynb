Задание 3 (1 балл)
Дан список:

nums = [3, 10, 5, 22, -4, 7, 0, 11]


Найдите:

- сумму чисел
- среднее число
- количество чисел больше среднего числа

```python
nums = [3, 10, 5, 22, -4, 7, 0, 11]
sum_nums = sum(nums)
print(f"Сумма чисел {sum_nums}")
avg = sum_nums / len(nums)
print(f"Среднее число {avg}")
col = 0
for n in nums:
    if n > avg:
        col +=1
print(f"Количество чисел больше среднего {col}")        
```

Задание 4 (1 балл)
Сумма продаж за день

Создайте функцию daily_sales(sales)
где sales — список продаж за день.
Если продажа > 15 000 ₽ — удерживается комиссия 10%. (нужно отнять 15% от продажи)
Функция возвращает общую выручку.

Пример:

sales = [16000, 7000, 1500, 23000, 17000]
print(daily_sales(sales))



```python
sales = [16000, 7000, 1500, 23000, 17000]

def daily_sales(sales):
    vir = 0
    for s in sales:
        if s > 15000:
            vir += s*0.85
        else:
            vir += s
    return vir
print(daily_sales(sales))               
```

Задание 5 (1 балл)
Дан словарь оценок:

grades = {"Sasha": 85, "Dima": 92, "Vova": 60, "Vlad": 74, "Alina": 100}


Найдите:

- среднюю оценку
- имя студента с максимальной оценкой
- имя студента с минимальной оценкой


```python
grades = {"Sasha": 85, "Dima": 92, "Vova": 60, "Vlad": 74, "Alina": 100}

avg = sum(grades.values()) / len (grades)
print(f"Средняя оценка {avg}")
max_g = max(grades, key=grades.get)
print(f"Студент с максимальной оценкой {max_g}")
min_g = min(grades, key=grades.get)
print(f"Студент с минимальной оценкой {min_g}")
```

Задание 9 (1 балл)
Найдите 10 городов (city_name) с самым высоким avg_income

То есть:
- Отсортировать по столбцу avg_income (по убыванию)
- Взять top-10 городов
- Построить bar chart, где:

Задание 10 (1 балл)

Oсь Y — название страны
Ось X — средний индекс качества воздуха

Задание 11 (2 балл)
Ваша задача сделать 2 графика на одном полотне (subplot)

- Первый график обычный линейный график, который показывает зависимость mean avg_rent и avg_income от density_range
  - По оси X - density_range (каждое значение 1 раз)
  - По оси Y - на графике должно быть две линии - mean avg_rent и avg_income 
  - Добавьте легенду на график
- Второй Pie Chart который показывает долю каждого значения density_range

```python
gr = df.groupby("density_range")[["avg_rent", "avg_income"]].mean().reset_index()
#print(gr)
dens = df["density_range"].value_counts()
#print(dens)

fig, (x1, x2) = plt.subplots(1, 2, figsize=(16, 8))

x1.plot(gr["density_range"], gr["avg_rent"], label="avg_rent", marker="o")
x1.plot(gr["density_range"], gr["avg_income"], label="avg_income", marker="s")
x1.set_xlabel("density_range")
x1.set_ylabel("avg_rent" "avg_income") 
x1.set_title("зависимость mean avg_rent и avg_income от density_range")
x1.legend()
x1.grid("--")

x2.pie(dens, labels=dens.index, autopct="%1.1f%%")
x2.set_title("Доля каждого значения density_range")
plt.tight_layout()
plt.show()
```
